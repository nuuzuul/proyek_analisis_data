## Import Semua Packages/Library yang Digunakan

Mengimpor library yang diperlukan untuk analisis data, yaitu numpy, pandas, matplotlib.pyplot, dan seaborn.*italicized text*

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data Wrangling

### Gathering Data

membaca data dari file CSV day.csv dan hour.csv dan menampilkan 5 baris pertama dari masing-masing dataset untuk melihat struktur datanya.

```python
# Membaca data dari file CSV
day_data = pd.read_csv("/content/sample_data/day.csv")
day_data.head()
```

```python
hour_data = pd.read_csv("/content/sample_data/hour.csv")
hour_data.head()
```

### Assessing Data

 memeriksa informasi dasar tentang dataset seperti tipe data dan jumlah data non-null, serta statistik deskriptif untuk memahami distribusi data.

```python
day_data.info()
hour_data.info()
```

```python
day_data.describe()
hour_data.describe()
```

mengecek apakah ada data yang duplikat dalam dataset untuk memastikan kualitas data.

```python
print("Jumlah duplikasi dalam data jam:", hour_data.duplicated().sum())
print("Jumlah duplikasi dalam data hari:", day_data.duplicated().sum())
```

### Cleaning Data

mengubah kolom tanggal dari tipe data string menjadi tipe data datetime untuk memudahkan analisis waktu.

```python
# Mengubah kolom tanggal menjadi tipe data datetime
datetime_columns = ["dteday"]
for column in datetime_columns:
    day_data[column] = pd.to_datetime(day_data[column])

for column in datetime_columns:
    hour_data[column] = pd.to_datetime(hour_data[column])
```

## Exploratory Data Analysis (EDA)

### Explore ... <br>



menghitung rata-rata penyewaan sepeda untuk setiap musim untuk memahami pola penyewaan berdasarkan musim.

```python
# Menghitung rata-rata penyewaan sepeda berdasarkan musim
rata_rata_musim = hour_data.groupby('season')['cnt'].mean()
print(rata_rata_musim)
```

menambahkan kolom baru yang menunjukkan nama hari dalam seminggu untuk analisis lebih lanjut berdasarkan hari.

```python
# Menambahkan kolom hari dalam seminggu
day_data['day_of_week'] = day_data['dteday'].dt.day_name()

# Menghitung rata-rata penyewaan sepeda berdasarkan hari
rata_rata_hari = day_data.groupby('day_of_week')['cnt'].mean()
print(rata_rata_hari)
```

## Visualization & Explanatory Analysis

Visualisasi Kinerja Penyewaan Sepeda Berdasarkan Musim

```python
rata_rata_musim.plot(kind='bar')
plt.xlabel('Musim')
plt.ylabel('Rata-rata Penyewaan Sepeda')
plt.title('Rata-rata Penyewaan Sepeda per Musim')
plt.xticks(range(4), ['Musim Semi', 'Musim Panas', 'Musim Gugur', 'Musim Dingin'])
plt.show()
```

Visualisasi Kinerja Penyewaan Sepeda Berdasarkan Hari

```python
plt.bar(rata_rata_hari.index, rata_rata_hari)
plt.xlabel('Hari dalam Seminggu')
plt.ylabel('Rata-rata Penyewaan Sepeda')
plt.title('Rata-rata Penyewaan Sepeda per Hari')
plt.xticks(rotation=45)
plt.show()
```

## Conclusion

- Kinerja Penyewaan Sepeda Berdasarkan Musim:

Berdasarkan analisis yang dilakukan, penyewaan sepeda tertinggi terjadi pada musim gugur (Fall), diikuti oleh musim panas (Summer) dan musim dingin (Winter).
Musim semi (Spring) menunjukkan tingkat penyewaan sepeda terendah. Hal ini mungkin disebabkan oleh kondisi cuaca yang lebih menguntungkan untuk bersepeda di musim gugur dibandingkan musim semi.

- Kinerja Penyewaan Sepeda Berdasarkan Hari:

Penyewaan sepeda paling banyak terjadi pada hari Jumat (Friday) dan Kamis (Thursday), yang menunjukkan tren penggunaan sepeda yang lebih tinggi menjelang akhir pekan.
Sebaliknya, hari dengan penyewaan sepeda paling sedikit adalah hari Minggu (Sunday). Ini bisa jadi karena hari Minggu adalah hari libur bagi banyak orang, yang mungkin memilih untuk beristirahat daripada beraktivitas di luar rumah.
